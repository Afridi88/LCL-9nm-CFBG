```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Load the data from a text file
path = os.getcwd()
data = np.loadtxt(path+'/rec.Pulse')

# Extract the columns from the data
time = data[:, 0]*1e-3
intensity = data[:, 1]
temporal_phase = data[:, 2]
chirp = data[:, 3]
wavelength = data[:, 4]
psd = data[:, 5]
spectral_phase = data[:, 6]
group_delay = data[:, 7]



# Create the first plot
fig1, ax1 = plt.subplots()
ax1.plot(time, intensity, 'b-', label='Intensity')
ax1.set_xlabel('Time (ps)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Intensity (arb. units)', fontsize=14, fontweight='bold')
ax1.set_xlim(-6,6)
ax1.tick_params(axis='both', which='major', labelsize=12)
ax1.tick_params(axis='both', which='minor', labelsize=10)
ax1.grid(which='major', linestyle='-', linewidth='0.5', color='black', alpha=0.6)
ax1.grid(which='minor', linestyle=':', linewidth='0.5', color='black', alpha=0.3)


ax2 = ax1.twinx()
ax2.plot(time, chirp, 'r-', label='Chirp')
ax2.set_ylim(-3000,3000)
ax2.set_ylabel('Chirp (GHz)', fontsize=14, fontweight='bold', color = 'red')
ax2.tick_params(axis='y', colors='red')
ax2.tick_params(axis='both', which='major', labelsize=12)
ax2.tick_params(axis='both', which='minor', labelsize=10)

plt.title('Pulse Time vs Intensity and Chirp', fontsize=18, fontweight='bold')

# Create the second plot
fig2, ax3 = plt.subplots()
ax3.plot(wavelength, psd, 'b-', label='PSD')
ax3.set_xlabel('Wavelength (nm)', fontsize=14, fontweight='bold')
ax3.set_ylabel('PSD (arb. units)', fontsize=14, fontweight='bold')
ax3.set_xlim(1020,1040)
ax3.set_xticks(np.linspace(1020,1040,5))
ax3.tick_params(axis='both', which='major', labelsize=12)
ax3.tick_params(axis='both', which='minor', labelsize=10)
ax3.grid(which='major', linestyle='-', linewidth='0.5', color='black', alpha=0.6)
ax3.grid(which='minor', linestyle=':', linewidth='0.5', color='black', alpha=0.3)

ax4 = ax3.twinx()
ax4.plot(wavelength, group_delay, 'r-', label='Group Delay')
ax4.set_ylabel('Group Delay (ps)', fontsize=14, fontweight='bold', color = 'red')
ax4.tick_params(axis='y', colors='red')
ax4.tick_params(axis='both', which='major', labelsize=12)
ax4.tick_params(axis='both', which='minor', labelsize=10)

           
plt.title('Wavelength vs PSD and Group Delay', fontsize=16, fontweight='bold')

plt.tight_layout()


# Show the plots
plt.show()


# fig1.savefig('/Users/nabb471/Desktop/DoddWalls2023/168ma200maChirp.png', bbox_inches='tight', pad_inches=0.2)
# fig2.savefig('/Users/nabb471/Desktop/DoddWalls2023/168ma200maGD.png', bbox_inches='tight', pad_inches=0.2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os



# Load the data from a text file
path = os.getcwd()
data = np.loadtxt(path+'/rec.Pulse')

# Read the data from the text file
with open(path+'/specto.Spectrogram', 'r') as file:
    lines = file.readlines()

# Extract the spectrogram configuration data from the first line
config_data = lines[0].split()
num_rows = int(config_data[0])
num_cols = int(config_data[1])
time_step = float(config_data[2])
space_step = float(config_data[3])
center_wavelength = float(config_data[4])

# Extract the spectrogram values from the remaining lines
spectrogram_data = []
for line in lines[1:]:
    spectrogram_data.extend([float(value) for value in line.split()])

# Reshape the spectrogram data into a 2D array
spectrogram_data = np.array(spectrogram_data).reshape(num_rows, num_cols)

# Normalize the spectrogram data to range between 0 and 1
spectrogram_data_normalized = (spectrogram_data - np.min(spectrogram_data)) / (np.max(spectrogram_data) - np.min(spectrogram_data))

# Create a meshgrid for the x and y coordinates
x = np.arange(num_cols)
y = np.arange(num_rows)

# Create a figure and plot the spectrogram
fig1, ax = plt.subplots()
im = ax.pcolormesh(x, y, spectrogram_data_normalized, shading='auto', cmap='jet')
ax.set_xlabel('Delay (ps)')
ax.set_ylabel('Wavelength (nm)')

# Customize the tick labels on the x and y axes
delay_ticks = np.linspace(0, num_cols - 1, 11)  # Specify the desired tick locations for the delay axis
wavelength_ticks = np.linspace(0, num_rows - 1, 11)  # Specify the desired tick locations for the wavelength axis
delay_tick_labels = np.linspace(-10, 10, 11)  # Specify the corresponding tick labels for the delay axis
wavelength_tick_labels = np.linspace(508, 528, 11)  # Specify the corresponding tick labels for the wavelength axis
ax.set_xticks(delay_ticks)
ax.set_yticks(wavelength_ticks)
ax.set_xticklabels([f'{value:.0f}' for value in delay_tick_labels])
ax.set_yticklabels([f'{value:.0f}' for value in wavelength_tick_labels])


ax.set_title('Spectrogram')
cbar = plt.colorbar(im, label='Intensity')
cbar.mappable.set_clim(0, 1)

# Display the plot
plt.show()


# path2savePlot = '/Users/nabb471/Desktop/linearcl/Frequently used/LCL modifications 2023/Recent data/LCL 18nm CFBG/FROG/Amp FROG at 168ma pump/plots'

# fig1.savefig(path2savePlot + '/amp200ma/168ma200maSpectogram.png', bbox_inches='tight', pad_inches=0.2)

# fig1.savefig('/Users/nabb471/Desktop/DoddWalls2023/168ma200maSpectogram.png', bbox_inches='tight', pad_inches=0.2)

```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Create a 2x2 grid with GridSpec, but use only three of the four cells
gs = GridSpec(2, 2)

# Create subplots with different configurations
ax1 = plt.subplot(gs[0, 0])
ax1.set_title('Plot 1')
ax1.plot([1, 2, 3], [1, 2, 3])

ax2 = plt.subplot(gs[0, 1])
ax2.set_title('Plot 2')
ax2.plot([1, 2, 3], [3, 2, 1])

# Create a subplot spanning two rows and two columns (a big subplot)
ax3 = plt.subplot(gs[1, :])
ax3.set_title('Plot 3')
ax3.plot([1, 2, 3], [1, 1, 1])

plt.tight_layout()
plt.show()
```
